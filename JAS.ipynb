```python

import matplotlib.backends.backend_pdf
import matplotlib.pyplot as plt
import random
import pylab as pl

root = "/home/pavel/Desktop/Work/Project/All_data/Worms/"
files_txt = sorted([f for f in os.listdir(root) if os.path.splitext(f)[1] == '.fasta'])

out_pdf = r'/home/pavel/Desktop/Work/Project/All_data/Worms.pdf'
pdf = matplotlib.backends.backend_pdf.PdfPages(out_pdf)


def mva(a, window):
    arr = []
    l = 0
    r = window
    for i in range(1, window + 1):
        #print(i)
        r = i * window
        s = a[l:r]
        #print(s)
        l = r
        arr.append(np.mean(s))
    return(arr)


figs = plt.figure()
os.chdir(root)


p = subprocess.Popen("rm *_al.*", shell=True)
p.wait()

for i in files_txt:
    tmp = ".".join(i.split(".")[:-1])
    print(tmp)
    try:
        line = 'clustalw -INFILE={in_file} -OUTFILE={out_file} -TYPE=DNA -OUTPUT=FASTA'.format(in_file = i, 
                                                                                              out_file = tmp + "_al.fasta")
        p = subprocess.Popen(line, shell=True)
        p.wait()
        
        line1 = 'snp-sites -mvp -o out.fasta {in_file}'.format(in_file = root + tmp + "_al.fasta")
        p = subprocess.Popen(line1, shell=True)
        p.wait()
                
        line2 = 'vcftools --hap-r2 --vcf out.fasta.vcf  --out out_vcftools'
        p = subprocess.Popen(line2, shell=True)
        p.wait()
           
        df = pd.read_csv(root + "out_vcftools.hap.ld", sep='\t',index_col=False)
        df["Len"] = df["POS2"] - df["POS1"]
    
        df = df.sort_values(by=['Len'])
        x = df["Len"]
        y = df["R^2"]
    
        window = len(x)//5
        a = mva(x, window)
        b = mva(y, window)
        
        f, ax = plt.subplots(figsize=(12, 8))
        #fig = sns.scatterplot(x="Len", y="R^2", data=df, ax=ax, s=80, marker="o")
        fig = sns.regplot(x=df["Len"], y=df["R^2"], fit_reg=False, scatter_kws={"color":"blue","alpha":0.1,"s":20} )
        fig = sns.lineplot(x="x", y="y", data=pd.DataFrame(data={"x":a, "y":b}), ax=ax, color="red")
        
        plt.title('R^2 plot. ' + tmp)
        plt.xlabel("Len")
        plt.ylabel("R^2")
        
        fig.set(ylim=(-0.05, 1.05))
        pdf.savefig()
        plt.clf()
        
        !rm out.fasta.vcf
        !rm out_vcftools.hap.ld
        !rm out_vcftools.log
        !rm out.fasta.phylip
        !rm out.fasta.snp_sites.aln
        
    except FileNotFoundError:
        continue
    #break
        
pdf.close()
```

```python

import matplotlib.backends.backend_pdf
import matplotlib.pyplot as plt
import random
import pylab as pl

circle_mode = True
aligned = True
name_format = "aligned"

root = "/home/pavel/Desktop/Work/Project/All_data/"


if name_format == "aligned_al":
    files_txt = sorted([f for f in os.listdir(root) if f.split(".")[-2:] == ['aligned_al', 'fasta']])
elif name_format == "aligned":
    files_txt = sorted([f for f in os.listdir(root) if f.split(".")[-2:] == ['aligned', 'fasta']])
else:
    files_txt = sorted([f for f in os.listdir(root) if f.split(".")[-1:] == ['fasta']])
out_pdf = r'/home/pavel/Desktop/Work/Project/All_data/Saccharomyces.pdf'
pdf = matplotlib.backends.backend_pdf.PdfPages(out_pdf)


def mva(a, window):
    arr = []
    l = 0
    r = window
    for i in range(1, window + 1):
        #print(i)
        r = i * window
        s = a[l:r]
        #print(s)
        l = r
        arr.append(np.mean(s))
    return(arr)
 

figs = plt.figure()
os.chdir(root)
for i in files_txt:
    
    with open(root + i, "r") as f:
        seqs={}
        for line in f:
            line = line.rstrip()
            if len(line) > 0:
                if line[0] == ">":
                    words = line.split()
                    name = words[0][1:]
                    seqs[name] = ''
                else:
                    seqs[name] = seqs[name] + line
        seq_len = len(list(seqs.values())[0])
    
    tmp = ".".join(i.split(".")[:-1])
    print(tmp)
    try:

        if aligned == False:
            line = 'clustalw -INFILE={in_file} -OUTFILE={out_file} -TYPE=DNA -OUTPUT=FASTA'.format(in_file = i, 
                                                                                              out_file = tmp + "_al.fasta")
            p = subprocess.Popen(line, shell=True)
            p.wait()
        
        
        line1 = 'snp-sites -mvp -o out.fasta {in_file}'.format(in_file = root + i)
        p = subprocess.Popen(line1, shell=True)
        p.wait()
                
        line2 = 'vcftools --hap-r2 --vcf out.fasta.vcf  --out out_vcftools'
        p = subprocess.Popen(line2, shell=True)
        p.wait()
           
        df = pd.read_csv(root + "out_vcftools.hap.ld", sep='\t',index_col=False)
        if len(list(df["POS1"])) > 100000:
            df1 = df.sample(100000)
        else:
            df1 = df
        if circle_mode == True:
            POS1 = list(df1["POS1"])
            POS2 = list(df1["POS2"])
            #print(POS1[:100])
            #print(POS2[:100])
            OUT = [None]*len(POS1)
            for line in range(len(POS1)):
                #print(line)
                #print(min(POS2[line] - POS1[line], seq_len - POS2[line] - POS1[line]))
                #print(POS2[line] - POS1[line])
                OUT[line] = min(abs(POS2[line] - POS1[line]), seq_len - abs(POS2[line] - POS1[line]))
            df1["Len"] = OUT
        else:
            df1["Len"] = df1["POS2"] - df1["POS1"]
    
        df1 = df1.sort_values(by=['Len'])
        x = df1["Len"]
        y = df1["R^2"]
    
        window = len(x)//100
        a = mva(x, window)
        b = mva(y, window)
        
        f, ax = plt.subplots(figsize=(12, 8))
        #fig = sns.scatterplot(x="Len", y="R^2", data=df, ax=ax, s=80, marker="o")
        fig = sns.regplot(x=df1["Len"], y=df1["R^2"], fit_reg=False, scatter_kws={"color":"blue","alpha":0.1,"s":20} )
        fig = sns.lineplot(x="x", y="y", data=pd.DataFrame(data={"x":a, "y":b}), ax=ax, color="red")
        
        plt.title('R^2 plot. ' + tmp)
        plt.xlabel("Len")
        plt.ylabel("R^2")
        
        fig.set(ylim=(-0.05, 1.05))
        pdf.savefig()
        plt.clf()
        
        !rm out.fasta.vcf
        !rm out_vcftools.hap.ld
        !rm out_vcftools.log
        !rm out.fasta.phylip
        !rm out.fasta.snp_sites.aln
        plt.close('all')
    except FileNotFoundError:
        continue
    print("done")
        
pdf.close()
```
